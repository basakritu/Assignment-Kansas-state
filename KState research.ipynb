# create frames from videos 

```python
def save_all_frames(video_path, dir_path, basename, ext='jpg'):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        return

    os.makedirs(dir_path, exist_ok=True)
    base_path = os.path.join(dir_path, basename)

    digit = len(str(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))))

    n = 0

    while True:
        ret, frame = cap.read()
        if ret:
            cv2.imwrite('{}_{}.{}'.format(base_path, str(n).zfill(digit), ext), frame)
            n += 1
        else:
            return

save_all_frames('pig_videos/Event20230919160512004.avi', 'pig_videos/Event20230919160512004/', 'sample_video_img')
```

# create the object detection model 

```python
class ObjectDetection:
    def __init__(self, weights_path="source_code/dnn_model/yolov3_training_last_latest.weights", cfg_path="source_code/dnn_model/yolov3_training.cfg"):
        print("Loading Object Detection")
        print("Running opencv dnn with YOLOv3")
        #Filter duplicate bounding bounding boxes above a certain threshold. 
        self.nmsThreshold = 0.4
        #Detect boxes with probability / confidence over a certain threshold
        self.confThreshold = 0.5
        self.image_size = 608

        # Load Network
        net = cv2.dnn.readNet(weights_path, cfg_path)

        # Enable GPU CUDA
        net.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
        net.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
        self.model = cv2.dnn_DetectionModel(net)

        self.colors = np.random.uniform(0, 255, size=(80, 3))

        self.model.setInputParams(size=(self.image_size, self.image_size), scale=1/255)


    def detect(self, frame):
        return self.model.detect(frame, nmsThreshold=self.nmsThreshold, confThreshold=self.confThreshold)

```

# create the animation for a given video with the bounding box:

# 0 - Toy, 1- Pig1 (with different ear tags), 2 - pig2 (same ear tags)

```python
# import cv2
# import numpy as np
# from object_detection import ObjectDetection
import math

# Initialize Object Detection
od = ObjectDetection()

# cap = cv2.VideoCapture("pig_videos/Event20230919145449004.avi")

cap = cv2.VideoCapture("pig_videos/Event20230919160512004.avi")

# Initialize count
count = 0
center_points_prev_frame = []

tracking_objects = {}
track_id = 0

start_frame_number = 0

while True:
    start_frame_number += 50
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame_number)
    ret, frame = cap.read()
    count += 1
    if not ret:
        break
    print(count)
    # Point current frame
    center_points_cur_frame = []

    # Detect objects on frame
    (class_ids, scores, boxes) = od.detect(frame)
    print("class_ids:", class_ids)
    print("scores:", scores)
    print("boxes:", boxes)
    for i, box in enumerate(boxes):
        (x, y, w, h) = box
        cx = int((x + x + w) / 2)
        cy = int((y + y + h) / 2)
        center_points_cur_frame.append((cx, cy, class_ids[i]))
        #print("FRAME N", count, " ", x, y, w, h)

        # cv2.circle(frame, (cx, cy), 5, (0, 0, 255), -1)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    #create the red point inside the box and the text
    for pt in center_points_cur_frame:
        point = (pt[0], pt[1])
        cv2.circle(frame, point, 5, (0, 0, 255), -1)
        cv2.putText(frame, str(pt[2]), (pt[0], pt[1] - 7), 0, 1, (0, 0, 255), 2)


    cv2.imshow("Frame", frame)

    # Make a copy of the points
    center_points_prev_frame = center_points_cur_frame.copy()

    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
